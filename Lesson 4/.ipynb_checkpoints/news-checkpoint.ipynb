```python
from lxml import html
import requests
from pprint import pprint
from datetime import date, timedelta
import pandas as pd
from pymongo import MongoClient
import json
import itertools
```

```python
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36'}
```

```python
def yandex_news():
    main_link = 'https://yandex.ru/news/'
    try:
        response = requests.get(main_link, headers=header)

        root = html.fromstring(response.text)
        items = root.xpath("//td[@class='stories-set__item'] | //div[@class='story story_view_main']")
        result = []
        for item in items:
            dict = {}
            source_date = str(item.xpath(".//div[@class='story__date']/text()")).replace(u'\\xa0в\\xa0', u' ').split(' ')

            if source_date[-2] != 'вчера':
                dict['date'] = f'{date.today()} {source_date[-1]}'
                dict['source'] = source_date[:-1]
            else: dict['date'] = f'{(date.today() - timedelta(days=1))} {source_date[-1]}'; \
                dict['source'] = source_date[:-2]

            dict['title'] = item.xpath(".//h2[@class='story__title']/*/text()")
            dict['link'] = item.xpath(".//h2[@class='story__title']/*/@href")
            result.append(dict)
    except Exception as e:
        print(e)

    yandex = pd.DataFrame(result)
    return yandex
```

```python
def lenta():
    main_link = 'https://lenta.ru'
    try:
        response = requests.get(main_link, headers=header)

        root = html.fromstring(response.text)
        items = root.xpath("//div[@class='first-item'] | //div[@class='item'] | //div[@class='item news b-tabloid__topic_news']")  #сознательно опущены статьи и новости внизу, не содержащие даты
        result = []
        for item in items:
            dict = {}
            news_date = item.xpath(".//time/@datetime | .//span[@class='g-date item__date']/text() | .//span[@class='g-date item__date']/span[1]/text()")
            today = str(date.today())
            if news_date == None:
                dict['date'] = today
            elif news_date[-1] == 'Сегодня':
                dict['date'] = news_date[0] + " " + today
            else: dict['date'] = news_date

            dict['title'] = str(item.xpath(".//a[@href]/text() | .//a[@href]/span/text()")).replace(u'\\xa0', u' ')
            dict['link'] = f'{main_link}{item.xpath(".//a[@href]/@href")[0]}'
            dict['source'] = 'Lenta.ru'
            result.append(dict)

    except Exception as e:
        print(e)
    lenta = pd.DataFrame(result)
    return lenta
```

```python
def news_mailru():
    main_link = 'https://news.mail.ru/'
    try:
        response = requests.get(main_link, headers=header)

        root = html.fromstring(response.text)
        items = root.xpath("//div[@class='daynews__item daynews__item_big'] | //div[@class='daynews__item'] | //li[@class='list__item'] | //li[@class='list__item hidden_small'] | //li[@class='list__item list__item_height_fixed_primary'] | //div[@class='newsitem newsitem_height_fixed newsitem_height_fixed_primary js-ago-wrapper']| //div[@class='newsitem newsitem_height_fixed js-ago-wrapper']")  #сознательно опущены справки, фоторепортажи, статьи

        result = []
        for item in items:
            dict = {}
            news_date = item.xpath(".//span[@class='newsitem__param js-ago']/@datetime")
            if news_date != []:
                dict['date'] = news_date
            else: dict['date'] = date.today()

            dict['title'] = str(item.xpath(".//a[@href]/text() | .//a[@href]/span/text() | .//span[@class='photo__title photo__title_new photo__title_new_hidden js-topnews__notification']/text()")).replace(u'\\xa0', u' ')
            dict['link'] = item.xpath(".//a[@href]/@href")

            source = item.xpath(".//span[@class='newsitem__param']/text()")

            if source != []:
                dict['source'] = source
            else: dict['source'] = 'News Mail.ru'
            result.append(dict)

    except Exception as e:
        print(e)
    mail = pd.DataFrame(result)
    return mail
```
